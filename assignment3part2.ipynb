```python
import random
import matplotlib.pyplot as plt

def hat_game():
    # Initialize the list with numbers from 1 to 100
    numbers = list(range(1, 11))

    # Continue the process until only one number is left
    while len(numbers) > 1:
        # Select two random numbers from the list
        index1 = random.randint(0, len(numbers) - 1)
        index2 = random.randint(0, len(numbers) - 1)

        # Ensure that the two indices are different
        while index1 == index2:
            index2 = random.randint(0, len(numbers) - 1)

        # Get the two numbers and compute their difference
        num1 = numbers[index1]
        num2 = numbers[index2]
        difference = abs(num1 - num2)

        # Remove the two chosen numbers from the list
        numbers.remove(num1)
        numbers.remove(num2)

        # Add the difference back to the list
        numbers.append(difference)

    # The while loop ends when there's only one number left
    final_number = numbers[0]
    return final_number

# Main function to run the game 100 times and plot the distribution
def main():
    final_numbers = []

    # Run the game 100,000 times
    for _ in range(100000):
        final_numbers.append(hat_game())

    # Count the frequency of each final number
    frequency = {}
    for number in final_numbers:
        frequency[number] = frequency.get(number, 0) + 1

    # Sort the final numbers and their frequencies
    final_numbers_sorted = sorted(frequency.keys())
    frequencies_sorted = [frequency[num] for num in final_numbers_sorted]

    # Plot the distribution of final numbers as a bar chart
    plt.bar(final_numbers_sorted, frequencies_sorted, align='center', color='skyblue')
    plt.title("Distribution of Final Numbers for n=11")
    plt.xlabel("Final Number")
    plt.ylabel("Frequency")
    plt.xticks(range(0, 11, 1))  # Adjust x-axis ticks for better readability
    plt.grid(axis='y')
    plt.show()

if __name__ == "__main__":
    main()
```
