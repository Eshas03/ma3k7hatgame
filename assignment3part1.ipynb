```python
import random
from collections import Counter

def hat_game():
    # Initialize the list with numbers from 1 to 100
    numbers = list(range(1, 2025))

    # Continue the process until only one number is left
    while len(numbers) > 1:
        # Select two random numbers from the list
        index1 = random.randint(0, len(numbers) - 1)
        index2 = random.randint(0, len(numbers) - 1)

        # Ensure that the two indices are different
        while index1 == index2:
            index2 = random.randint(0, len(numbers) - 1)

        # Get the two numbers and compute their difference
        num1 = numbers[index1]
        num2 = numbers[index2]
        difference = abs(num1 - num2)

        # Remove the two chosen numbers from the list
        numbers.remove(num1)
        numbers.remove(num2)

        # Add the difference back to the list
        numbers.append(difference)

    # The while loop ends when there's only one number left
    final_number = numbers[0]
    return final_number

# Main function to run the game multiple times and print distribution
def main():
    # Run the game 100 times
    results = []
    for _ in range(100):
        final_number = hat_game()
        results.append(final_number)
    
    # Count the occurrences of each final number
    distribution = Counter(results)

    # Print the distribution
    print("Distribution of final numbers for n=2024:")
    for number, count in sorted(distribution.items()):
        print(f"Number {number}: {count} occurrences")

if __name__ == "__main__":
    main()
```
